**Per-exercise source tables**

One source clone per exercise so streams cannot pollute each other. Each exN_source is a Delta CTAS from db_code.streaming.events_source.

```python
# exercise 1, 4,6 

for ex in [1,2,3,4,5,6,7]:
    spark.sql(f"""CREATE OR REPLACE TABLE {catalog}.{schema}.ex{ex}_source
    USING DELTA 
    AS SELECT * FROM {catalog}.{Base_schema}.event_source""")
```
